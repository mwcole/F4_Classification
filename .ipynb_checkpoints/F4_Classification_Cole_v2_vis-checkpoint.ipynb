# Support vector machine example

## Set up and run support vector machine (SVM) on example data

```python
from sklearn.svm import SVC

#Set data variables
X=example_data_X
y=example_data_y

# Train a linear SVM classifier on the 3D data
clf = SVC(kernel='linear')
clf.fit(X, y)
```

## Plot the SVM decision hyperplane with example data

```python
import plotly.graph_objects as go
import math

# Extract the oefficients and intercept from the trained SVM
w = clf.coef_[0]
b = clf.intercept_[0]

# For the decision plane defined by: w[0]*x + w[1]*y + w[2]*z + b = 0,
# solve for z: z = -(w[0]*x + w[1]*y + b) / w[2]
# Create a mesh grid over the range of x and y values from the data
xx_range = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 10)
yy_range = np.linspace(X[:, 1].min()-1, X[:, 1].max()+1, 10)
xx, yy = np.meshgrid(xx_range, yy_range)

# Compute the corresponding z values on the decision plane
if abs(w[2]) > 1e-6:
    zz = -(w[0] * xx + w[1] * yy + b) / w[2]
else:
    # In case w[2] is zero, set z as the mean of the z values (a vertical plane)
    zz = np.full_like(xx, X[:, 2].mean())

# Create scatter traces for the two classes with different colors
scatter_class0 = go.Scatter3d(
    x=X[y == 0, 0],
    y=X[y == 0, 1],
    z=X[y == 0, 2],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Class 0'
)

scatter_class1 = go.Scatter3d(
    x=X[y == 1, 0],
    y=X[y == 1, 1],
    z=X[y == 1, 2],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Class 1'
)

# Create a surface trace for the SVM decision plane
plane_surface = go.Surface(
    x=xx,
    y=yy,
    z=zz,
    colorscale='Viridis',
    opacity=0.5,
    name='Decision Plane',
    showscale=False  # This disables the colorbar
)

#Improve initial camera angle
# Define spherical coordinates for the camera
# mpl_toolkits.mplot3d default view typically uses:
#   elev = 30° and azim = -60°
r = 2.5  # distance of the camera from the center (you can adjust this)
elev = math.radians(30)
azim = math.radians(-60)
# Convert spherical to Cartesian coordinates
eye_x = r * math.cos(elev) * math.cos(azim)
eye_y = r * math.cos(elev) * math.sin(azim)
eye_z = r * math.sin(elev)

# Combine the traces into one figure and update the layout
fig = go.Figure(data=[scatter_class0, scatter_class1, plane_surface])
fig.update_layout(
    title='3D Scatter Plot with SVM Decision Plane',
    scene_camera=dict(
        eye=dict(x=eye_x, y=eye_y, z=eye_z)
    ),
    scene=dict(
        xaxis_title='Feature 1',
        yaxis_title='Feature 2',
        zaxis_title='Feature 3',
        aspectmode='cube'  # Forces equal scaling on all axes
    ),
    width=1000,       # width in pixels
    height=500,       # height in pixels
    margin=dict(l=40, r=50, t=40, b=40),  # adjust margins as needed
    
)

# Display the interactive plot
fig.show()
```

# Minimum distance classifier example

## Set up and run minimum distance classifier

```python
#Set data variables
X=example_data_X
y=example_data_y

# Compute class means
mean0 = np.mean(X[y == 0], axis=0)
mean1 = np.mean(X[y == 1], axis=0)
# Define a common center for centering (global midpoint of the means)
center = (mean0 + mean1) / 2.0

# For our modified Pearson/cosine similarity,
# the decision boundary is the set of x where (x - center) is orthogonal to (mean1 - center).
d = mean1 - center  # note that (mean0 - center) = -d
# The hyperplane is given by (x - center)·d = 0.
```

## Plot the minimum distance classifier decision boundary

```python

# To plot this hyperplane as a surface, solve for z in terms of x and y.
# That is, if d = [d_x, d_y, d_z], then:
#   d_x*(x - center_x) + d_y*(y - center_y) + d_z*(z - center_z) = 0.
# If d_z is not zero, solve:
#   z = center_z - (d_x*(x - center_x) + d_y*(y - center_y)) / d_z

x_vals = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 15)
y_vals = np.linspace(X[:, 1].min()-1, X[:, 1].max()+1, 15)
xx, yy = np.meshgrid(x_vals, y_vals)

if abs(d[2]) > 1e-6:
    zz = center[2] - (d[0]*(xx - center[0]) + d[1]*(yy - center[1])) / d[2]
else:
    zz = np.full_like(xx, center[2])

# Create scatter traces for the two classes
scatter_class0 = go.Scatter3d(
    x=X[y == 0, 0],
    y=X[y == 0, 1],
    z=X[y == 0, 2],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Class 0'
)

scatter_class1 = go.Scatter3d(
    x=X[y == 1, 0],
    y=X[y == 1, 1],
    z=X[y == 1, 2],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Class 1'
)

# Optionally, plot the class means for reference
mean0_trace = go.Scatter3d(
    x=[mean0[0]],
    y=[mean0[1]],
    z=[mean0[2]],
    mode='markers',
    marker=dict(size=8, color='cyan', symbol='diamond'),
    name='Mean Class 0'
)
mean1_trace = go.Scatter3d(
    x=[mean1[0]],
    y=[mean1[1]],
    z=[mean1[2]],
    mode='markers',
    marker=dict(size=8, color='magenta', symbol='diamond'),
    name='Mean Class 1'
)

# Plot the decision boundary as a surface
plane = go.Surface(
    x=xx,
    y=yy,
    z=zz,
    colorscale='Viridis',
    opacity=0.5,
    name='Decision Boundary',
    showscale=False  # This disables the colorbar
)

#Improve initial camera angle
# Define spherical coordinates for the camera
# mpl_toolkits.mplot3d default view typically uses:
#   elev = 30° and azim = -60°
r = 2.5  # distance of the camera from the center (you can adjust this)
elev = math.radians(30)
azim = math.radians(-60)
# Convert spherical to Cartesian coordinates
eye_x = r * math.cos(elev) * math.cos(azim)
eye_y = r * math.cos(elev) * math.sin(azim)
eye_z = r * math.sin(elev)

# Combine all traces into one figure
fig = go.Figure(data=[scatter_class0, scatter_class1, plane, mean0_trace, mean1_trace])
fig.update_layout(
    title='3D Minimum Distance Classifier (Modified Pearson/ Cosine Similarity)',
    scene_camera=dict(
        eye=dict(x=eye_x, y=eye_y, z=eye_z)
    ),
    scene=dict(
        xaxis_title='Feature 1',
        yaxis_title='Feature 2',
        zaxis_title='Feature 3',
        aspectmode='cube'  # Forces equal scaling on all axes
    ),
    width=1000,
    height=500,
    margin=dict(l=50, r=50, t=50, b=50)
)

fig.show()
```

# Neural network classification example

## Set up and run neural network model

```python
from sklearn.neural_network import MLPClassifier

# Train a simple neural network classifier (MLP) on the 3D data
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42)
mlp.fit(X, y)
```

## Plot the neural network decision boundary

```python

# Define the decision function using the difference between predicted probabilities:
# f(x) = P(class=1) - P(class=0)
def decision_function(points):
    # 'points' should be of shape (n_samples, n_features)
    probs = mlp.predict_proba(points)
    return probs[:, 1] - probs[:, 0]

# Create a 3D grid over the data range to evaluate the decision function
num_points = 20  # grid resolution along each axis
x_lin = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, num_points)
y_lin = np.linspace(X[:, 1].min()-1, X[:, 1].max()+1, num_points)
z_lin = np.linspace(X[:, 2].min()-1, X[:, 2].max()+1, num_points)
XX, YY, ZZ = np.meshgrid(x_lin, y_lin, z_lin)
grid_points = np.vstack([XX.ravel(), YY.ravel(), ZZ.ravel()]).T

# Evaluate the decision function on each grid point
f_values = decision_function(grid_points)

# Plot the decision boundary using an isosurface at f(x)=0.
# Adjust isomin/isomax to a small interval around zero.
isosurface = go.Isosurface(
    x=grid_points[:, 0],
    y=grid_points[:, 1],
    z=grid_points[:, 2],
    value=f_values,
    isomin=-0.01,   # tolerance around 0
    isomax=0.01,
    surface_count=1,
    colorscale='gray',
    opacity=0.5,
    name='Decision Boundary',
    showscale=False  # This disables the colorbar
)

# Create scatter traces for the two classes
scatter_class0 = go.Scatter3d(
    x=X[y == 0, 0],
    y=X[y == 0, 1],
    z=X[y == 0, 2],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='Class 0'
)
scatter_class1 = go.Scatter3d(
    x=X[y == 1, 0],
    y=X[y == 1, 1],
    z=X[y == 1, 2],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Class 1'
)

# Optionally, plot the class means for reference
mean0 = np.mean(X[y == 0], axis=0)
mean1 = np.mean(X[y == 1], axis=0)
mean0_trace = go.Scatter3d(
    x=[mean0[0]],
    y=[mean0[1]],
    z=[mean0[2]],
    mode='markers',
    marker=dict(size=8, color='cyan', symbol='diamond'),
    name='Mean Class 0'
)
mean1_trace = go.Scatter3d(
    x=[mean1[0]],
    y=[mean1[1]],
    z=[mean1[2]],
    mode='markers',
    marker=dict(size=8, color='magenta', symbol='diamond'),
    name='Mean Class 1'
)

#Improve initial camera angle
# Define spherical coordinates for the camera
# mpl_toolkits.mplot3d default view typically uses:
#   elev = 30° and azim = -60°
r = 2.5  # distance of the camera from the center (you can adjust this)
elev = math.radians(30)
azim = math.radians(-60)
# Convert spherical to Cartesian coordinates
eye_x = r * math.cos(elev) * math.cos(azim)
eye_y = r * math.cos(elev) * math.sin(azim)
eye_z = r * math.sin(elev)

# Combine all traces into one figure
fig = go.Figure(data=[scatter_class0, scatter_class1, isosurface, mean0_trace, mean1_trace])
fig.update_layout(
    title='3D Neural Network Classifier Decision Boundary',
    scene_camera=dict(
        eye=dict(x=eye_x, y=eye_y, z=eye_z)
    ),
    scene=dict(
        xaxis_title='Feature 1',
        yaxis_title='Feature 2',
        zaxis_title='Feature 3',
        aspectmode='cube'  # Forces equal scaling on all axes
    ),
    width=1000,
    height=500,
    margin=dict(l=50, r=50, t=50, b=50)
)

fig.show()
```

```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='Neural Network Architecture', format='svg')

# Set layout attributes for left-to-right orientation
dot.attr(rankdir='LR', splines='line', nodesep='0.5', ranksep='0.75')

# Input layer (3 nodes)
with dot.subgraph(name='cluster_input') as c:
    c.attr(rank='same')
    for i in range(1, 4):
        c.node(f'I{i}', shape='circle', style='filled', fillcolor='lightblue', label=f'Input {i}')

# Hidden layer (10 nodes)
with dot.subgraph(name='cluster_hidden') as c:
    c.attr(rank='same')
    for i in range(1, 11):
        c.node(f'H{i}', shape='circle', style='filled', fillcolor='lightgreen', label=f'H{i}')

# Output layer (2 nodes)
with dot.subgraph(name='cluster_output') as c:
    c.attr(rank='same')
    for i in range(1, 3):
        c.node(f'O{i}', shape='circle', style='filled', fillcolor='lightpink', label=f'Output {i}')

# Connect every input node to every hidden node
for i in range(1, 4):
    for j in range(1, 11):
        dot.edge(f'I{i}', f'H{j}')

# Connect every hidden node to every output node
for i in range(1, 11):
    for j in range(1, 3):
        dot.edge(f'H{i}', f'O{j}')

# Render the graph to an SVG file (this file can be imported into Adobe Illustrator)
dot.render('neural_network_architecture', view=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Create a range of x values
x = np.linspace(-10, 10, 400)
y = relu(x)

# Create the plot
plt.figure(figsize=(3, 2))
plt.plot(x, y, linewidth=2, color='blue', label='ReLU: f(x) = max(0, x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Rectified Linear Unit (ReLU) Activation Function')
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.axvline(0, color='black', linewidth=1, linestyle='--')
plt.legend()
plt.grid(True)

# Save the plot as an SVG file for Adobe Illustrator
plt.savefig('relu_function.svg', format='svg')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(-10, 10, 400)

# Compute softmax outputs for the vector [x, 0]
p1 = np.exp(x) / (np.exp(x) + 1)  # Probability for the first neuron
p0 = 1 / (np.exp(x) + 1)            # Probability for the second neuron

# Create the plot
plt.figure(figsize=(3, 2))
plt.plot(x, p1, linewidth=2, color='red', label=r'$P(y=1)=\frac{e^x}{e^x+1}$')
plt.plot(x, p0, linewidth=2, color='blue', label=r'$P(y=0)=\frac{1}{e^x+1}$')
plt.xlabel(r'$x$ (input difference)')
plt.ylabel('Probability')
plt.title('Softmax Function for Two Outputs: [x, 0]')
plt.axvline(0, color='black', linestyle='--', linewidth=1)
plt.axhline(0.5, color='black', linestyle='--', linewidth=1)
plt.legend()
plt.grid(True)

# Save the plot as an SVG file for vector editing in Adobe Illustrator
plt.savefig('softmax_function.svg', format='svg')
plt.show()
```
